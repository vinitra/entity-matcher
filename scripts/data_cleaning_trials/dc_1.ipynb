```python
import pandas as pd
import numpy as np
```

```python
df = pd.read_csv('../../data/X2.csv')
```

```python
pattern_string = '( - )|,|:|\(|\)'
preprocessed_titles = df["title"] \
    .apply(lambda s: s.strip().lower()) \
    .replace(pattern_string, ' ', regex=True)

df.insert(len(df.columns), "preprocessed_titles", preprocessed_titles)
```

```python
df.iloc[:5,-2:]
```

```python
titles = df["preprocessed_titles"].values
titles.shape
```

```python
from sklearn.feature_extraction.text import TfidfVectorizer
```

```python
vectoriser = TfidfVectorizer(stop_words='english', ngram_range=(1, 2), token_pattern='\\b(\\w+(?:[\\.-]\\w+)*)\\b')

#token_pattern="(?u)\\b\\w[-\\w]*\\b")
```

```python
X = vectoriser.fit_transform(titles)
```

```python
X.shape
```

```python
feature_names = vectoriser.get_feature_names()
feature_names
```

```python
n = 66
feature_indexes = X[n, :].nonzero()[1]
print(len(feature_indexes))
tf_idf_scores = [(X[n, x], x) for x in feature_indexes]
tf_idf_scores.sort(reverse=True)
sz = len(feature_indexes)//2
for s, i in tf_idf_scores[:]:
    print(feature_names[i], s)
```
