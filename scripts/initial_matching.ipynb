```python
import pandas as pd
import numpy as np
import os
from fuzzymatcher.data_preprocessor_default import DataPreprocessor
from fuzzymatcher.data_getter_sqlite import DataGetter
from fuzzymatcher.scorer_default import Scorer
from fuzzymatcher.matcher import Matcher

# load a dataframe

df = x1_raw
match_cols = df.columns.drop('instance_id')

# split into a left and right copy
df_left = df[['instance_id', 'cpu_model']].copy()
df_right = df[['instance_id', 'cpu_model']].copy()

# create a match object
dp = DataPreprocessor()
dg = DataGetter()
s = Scorer()
m = Matcher(dp, dg, s)

# match the data with a copy of itself
for col in match_cols:
    m.add_data(df_left, df_right, left_on=col, right_on=col, left_id_col='instance_id', right_id_col='instance_id')
m.match_all()
df_duplicated = m.link_table

# subset the data where the ids don't match - these are the duplicates
df_duplicated = df_duplicated[df_duplicated.__id_left != df_duplicated.__id_right].copy()
```

```python
best_matches = df_duplicated.dropna(0, "any").sort_values(by="__score", ascending=False)
best_matches = best_matches[best_matches['__score'] != -float("inf")]
best_matches
```
